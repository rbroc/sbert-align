4869 and 9611 have repeated turns for Visit 1. Solution: remove the first of the two.

```python
df[(df['ID']==9611) & (df['Visit']==1)].sort_values(by='Turn').head(n=50)[['Turn', 'Transcript', 'StartTime', 'EndTime', 'Speaker']]
```

```python
df.sort_values(by=['ID', 'Visit', 'Task', 'Turn']).groupby(['ID', 'Visit', 'Task']).apply(lambda x: x[x['Speaker']==x.shift()['Speaker']][['ID', 'Visit', 'Task', 'Turn', 'Speaker', 'StartTime', 'EndTime', 'Transcript']])
```

```python
# 4869 and 9611 are taken care of
```

```python
df.sort_values(by=['ID', 'Visit', 'Task', 'Turn']).groupby(['ID', 'Visit', 'Task']).apply(lambda x: x[x['Turn']-x.shift()['Turn']>1][['ID', 'Visit', 'Task', 'Turn', 'Speaker', 'StartTime', 'EndTime', 'Transcript']])
```

```python
df[(df['ID']==9035) & (df['Visit']==1) & (df['Task']=='MatchingGame')].sort_values(by='Turn')[['ID', 'Visit', 'Task', 'Turn', 'Speaker', 'StartTime', 'EndTime', 'Transcript']]
# jumping from 111 to 114 (then also jumping more)
```

```python
no_current = df.sort_values(by=['ID', 'Visit', 'Task', 'Turn']).groupby(['ID', 'Visit', 'Task'], as_index=False).apply(lambda x: x[x['Turn']-x.shift()['Turn']>1])[['ID', 'Visit', 'Task', 'Turn']].apply(lambda row: '_'.join(row.values.astype(str)), axis=1).values.tolist()
```

```python
no_current, df['unique_id']
```

```python
df['unique_id'] = df[['ID', 'Visit', 'Task', 'Turn']].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
```

```python
df['unique_id']
```

```python
df['no_previous'] = df['unique_id'].apply(lambda x: f"{'_'.join(x.split('_')[:3])}_{int(x.split('_')[-1])-1}" in no_current)
```

```python
no_previous = df[df['no_previous']==True].unique_id.tolist()
```

```python
no_current # make both nan
```

```python
import json 

exclude = {'alignment_current_to_1back': no_current, 'alignment_1back_to_2back': no_previous + no_current}
with open('sbert-align/data/exclude.json', 'w') as f:
    json.dump(exclude, f)
```
